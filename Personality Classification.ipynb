```python
import numpy as np
numeric_data = df.select_dtypes(include=[np.number])
z_scores = (numeric_data - numeric_data.mean()) / numeric_data.std()
threshold = 3
outliers_mask = (abs(z_scores) > threshold)
outlier_counts = outliers_mask.sum()
print("Outlier count per column:")
print(outlier_counts)
```

```python
total_outliers = outliers_mask.sum().sum()
total_values = numeric_data.size
outlier_percentage = (total_outliers / total_values) * 100

print(f"Total number of outliers: {total_outliers}")
print(f"Percentage of outliers in numeric data: {outlier_percentage:.2f}%")
```
