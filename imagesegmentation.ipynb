```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
# Load image and convert to array
img = Image.open(r"C:\Users\gargn\Desktop\ds\Screenshot 2025-07-11 000505.png")
a = np.array(img)
# Create a single RGBA overlay (initialized to zeros)
overlay = np.zeros((a.shape[0], a.shape[1], 4), dtype=np.uint8)
# Brightness threshold
th =110
# Define your 4 quadrant boxes and colors (RGBA)
quadrants = [
    # top-left box      color=Cyan
    ((30, 110,  85, 130), [  0, 255, 255, 120]),
    # top-right box     color=Red
    ((30, 110, 130, 168), [255,   0,   0, 120]),
    # bottom-left box   color=Magenta
    ((110, 152,  105, 130), [255,   0, 255, 120]),
    ((110, 145,  100, 130), [255,   0, 255, 120]),
    ((110, 140,  97, 130), [255,   0, 255, 120]),
    ((110, 133,  92, 130), [255,   0, 255, 120]),
    # bottom-right box  color=Green
    ((110, 153, 130, 148), [  0, 255,   0, 120]),
    ((110, 145, 130, 155), [  0, 255,   0, 120]),
    ((110, 140, 130, 159), [  0, 255,   0, 120]),
    ((110, 132, 130, 162), [  0, 255,   0, 120]),
    ((110, 128, 130, 164), [  0, 255,   0, 120]),
]
# Paint each quadrant into the same overlay
for (r0, r1, c0, c1), color in quadrants:
    for i in range(r0, r1):
        for j in range(c0, c1):
            if np.mean(a[i, j]) > th:
                overlay[i, j] = color
# Display: base image first, then the combined translucent overlay
plt.imshow(a)
plt.imshow(overlay, alpha=overlay[..., 3] / 255)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load image
img = Image.open(r"C:\Users\gargn\Downloads\photo-1542909168-82c3e7fdca5c.jpg")
a = np.array(img)

# Face bounding box - update if needed
start_row, end_row = 1000, 2000
start_col, end_col = 1225, 2000

# Loop through region and apply color only to skin-tone pixels
for i in range(start_row, end_row):
    for j in range(start_col, end_col):
        pixel = a[i, j]
        r, g, b = pixel

        # Simple face/skin detection condition (tuned manually)
        if r > 80 and g > 40 and b > 20 and (max(pixel) - min(pixel)) < 60:
            a[i, j] = [255, 255, 0]  # Yellow color

# Show result
plt.imshow(a)
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load image and convert to array
img = Image.open(r"C:\Users\gargn\Downloads\photo-1542909168-82c3e7fdca5c.jpg")
a = np.array(img)

# Define bounding box for the face
start_row, end_row = 900,2500
start_col, end_col = 1225, 2500

# Calculate midpoint of face region
mid_row = (start_row + end_row) // 2
mid_col = (start_col + end_col) // 2

# Apply color using two loops with quadrant logic
for i in range(start_row, end_row):
    for j in range(start_col, end_col):
        if np.mean(a[i, j]) > 35:  # Brightness threshold for skin
            if i < mid_row and j < mid_col:
                a[i, j] = [0, 255, 255]    # Cyan - top-left
            elif i < mid_row and j >= mid_col:
                a[i, j] = [255, 255, 0]    # Yellow - top-right
            elif i >= mid_row and j < mid_col:
                a[i, j] = [255, 1, 255]    # Magenta - bottom-left
            else:
                a[i, j] = [0, 255, 0]      # Green - bottom-right

# Show the modified image
plt.imshow(a)
# plt.axis('off')
# plt.show()
```
